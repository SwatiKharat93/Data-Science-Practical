1.Explain the role of NumPy in the Python scientific computing ecosystem.

NumPy plays a fundamental role in the Python scientific computing ecosystem by providing powerful tools for numerical operations and data manipulation. It introduces the concept of multidimensional arrays, which are faster and more efficient than Python lists. NumPy serves as the foundation for many other libraries used in data science, machine learning, and scientific research. It enables users to perform complex mathematical computations, handle large datasets, and implement algorithms efficiently, making it essential for scientific computing.

2.Why is NumPy considered faster than pure Python computations?

NumPy is considered faster than pure Python computations because it uses optimized C language code internally, which executes operations more efficiently. Unlike Python lists that require loops for calculations, NumPy performs operations on entire arrays at once using vectorization. This reduces the need for explicit loops and improves speed. Additionally, NumPy stores data in contiguous memory locations, which enhances performance and reduces memory usage, making it significantly faster for numerical tasks.

3.What types of mathematical operations can be performed using NumPy?

NumPy supports a wide range of mathematical operations, including basic arithmetic operations such as addition, subtraction, multiplication, and division. It also provides functions for advanced computations like mean, median, standard deviation, variance, and correlation. NumPy can perform matrix operations such as matrix multiplication, transpose, and determinant calculation. Additionally, it includes functions for trigonometry, logarithms, exponentials, and statistical analysis, making it suitable for scientific and engineering applications.

4.Explain indexing and slicing in NumPy arrays.

Indexing and slicing in NumPy arrays are used to access and manipulate specific elements or portions of the data. Indexing allows users to retrieve individual elements by specifying their position using row and column indices. Slicing enables accessing a subset of the array by specifying a range of indices. For example, users can extract specific rows, columns, or a block of data from a multidimensional array. These features make data handling flexible and efficient, especially when working with large datasets.

5.How does NumPy integrate with other libraries?
a. Pandas

NumPy integrates closely with the Pandas library, as Pandas uses NumPy arrays as its underlying data structure. DataFrames and Series in Pandas are built on NumPy, which allows fast data manipulation and numerical operations.

b. Matplotlib

NumPy works with Matplotlib to create visualizations. NumPy arrays are often used as input data for plotting graphs such as line charts, bar charts, and histograms, making data visualization easy and efficient.

c. Scikit-learn

NumPy integrates with Scikit-learn for machine learning tasks. Most algorithms in Scikit-learn require input data in the form of NumPy arrays, enabling efficient computation and model training.

```python
import numpy as np
```

```python
data = np.array([
    [78, 85, 90, 88, 76],
    [65, 70, 72, 68, 66],
    [90, 92, 94, 89, 91],
    [55, 60, 58, 62, 57],
    [88, 84, 86, 90, 87],
    [76, 79, 75, 80, 78],
    [92, 95, 93, 96, 94],
    [60, 65, 63, 67, 62],
    [82, 80, 85, 83, 81],
    [70, 72, 68, 74, 69]
])
```

```python
print("Dataset" , data)
```

```python
sorted_data = np.sort(data)
print(sorted_data)
```

```python
print("Total Marks:", np.sum(data))
print("Average Marks:", np.mean(data))
print("Maximum Marks:", np.max(data))
print("Minimum Marks:", np.min(data))
```

```python
total_per_student = np.sum(data, axis=1)
print("Total per Student:", total_per_student)
```

```python
avg_per_subject = np.mean(data, axis=0)
print("Average per Subject:", avg_per_subject)
```

```python
print("Standard Deviation:", np.std(data))
```

```python
high_marks = data[data > 80]
print("Marks > 80:", high_marks)
```

```python
# Add 5 grace marks
new_data = data + 5
print("After Adding Grace Marks:\n", new_data)
```

```python
print("Square:\n", np.square(data))
print("Square Root:\n", np.sqrt(data))
```

```python
# Addition
result_add = data + data
print(result_add)

# Multiplication
result_mul = np.dot(data, data.T)
print(result_mul)
```

```python
flat_data = data.flatten()
```

```python
Q1 = np.percentile(flat_data, 25)
Q2 = np.percentile(flat_data, 50)
Q3 = np.percentile(flat_data, 75)

print("Q1:", Q1)
print("Median (Q2):", Q2)
print("Q3:", Q3)
```

```python
# indexing
print(data[0, 2])   # First student, 3rd subject
```

```python
# Slicing
print(data[0:3])        # First 3 students
print(data[:, 1])       # All students, 2nd subject
```

```python
# Using the cloud Dataset
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("cloud_outages_dataset.csv")
df.head()
```

```python
print("Shape of dataset:", df.shape)  # Check Data
print("\nColumns:\n", df.columns)
```

```python
numeric_df = df.select_dtypes(include=[np.number])  # Select Only Numeric Columns
numeric_df.head()
```

```python
data = numeric_df.to_numpy()
print(data[:5])   # first 5 rows
```

```python
mean_values = np.mean(data, axis=0)
print("Mean:\n", mean_values)
```

```python
median_values = np.median(data, axis=0)
print("Median:\n", median_values)
```

```python
std_values = np.std(data, axis=0)
print("Standard Deviation:\n", std_values)
```

```python
print("Minimum values:\n", np.min(data, axis=0))
print("Maximum values:\n", np.max(data, axis=0))
```

```python
print("Sum of each column:\n", np.sum(data, axis=0))
```

```python
transpose_data = data.T
print("Transpose (first 5 values):\n", transpose_data[:, :5])
```

```python
correlation = np.corrcoef(data, rowvar=False)
print("Correlation Matrix:\n", correlation)
```
