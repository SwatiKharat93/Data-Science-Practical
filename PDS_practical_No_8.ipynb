1.What is correlation? How is it different from regression?
Correlation is a statistical method used to measure the strength and direction of the relationship between two variables. It tells us whether variables move together (positive correlation), move in opposite directions (negative correlation), or have no relationship. However, correlation does not imply cause-and-effect; it only shows association.
Regression, on the other hand, is used to predict the value of one variable based on another. It establishes a functional relationship between dependent and independent variables. While correlation gives a single value (r), regression provides an equation (like Y = a + bX) that helps in prediction. In short, correlation measures relationship strength, whereas regression focuses on prediction and modeling.

2.What is simple linear regression? Explain with suitable example.
Simple linear regression is a statistical technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line. The equation is:
Y = a + bX
where a is the intercept and b is the slope of the line.
For example, consider study hours (X) and exam marks (Y). If regression analysis gives the equation Y = 20 + 8X, it means that for every additional hour of study, marks increase by 8. If a student studies 5 hours, predicted marks = 20 + (8×5) = 60. This method helps in forecasting outcomes based on input values.

3.How do you calculate Pearson correlation in Python?
Pearson correlation in Python can be calculated using libraries like NumPy. The corrcoef() function is commonly used. First, we define two arrays for variables X and Y. Then we apply np.corrcoef(X, Y) which returns a matrix, and the correlation value is obtained from it. For example:
import numpy as np

X = [1, 2, 3, 4, 5]
Y = [30, 35, 50, 55, 65]

r = np.corrcoef(X, Y)[0, 1]
print(r)

4.What is multicollinearity? Does it affect simple regression?
Multicollinearity occurs when two or more independent variables are highly correlated with each other in a regression model. It creates problems in estimating the individual effect of each variable because they provide similar information.

In simple linear regression, there is only one independent variable, so multicollinearity does not occur. It is mainly a problem in multiple regression models where multiple predictors are used. Therefore, simple regression is not affected by multicollinearity.


5.A dataset shows strong correlation but poor prediction accuracy. Why?
A dataset may show strong correlation but still result in poor prediction accuracy due to several reasons. Correlation only measures the strength of a linear relationship and does not guarantee that the relationship is useful for prediction. The data may contain outliers, noise, or non-linear patterns that reduce model accuracy. Additionally, overfitting or underfitting of the regression model can lead to poor predictions. Another reason could be that important variables affecting the outcome are missing, so even though two variables appear strongly related, they do not fully explain the behavior of the dependent variable. Hence, high correlation does not always ensure accurate predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample Data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([30, 35, 50, 55, 65])
```

```python
r = np.corrcoef(X, Y)[0, 1]
print("Karl Pearson Correlation Coefficient (r):", r)
```

```python
n = len(X)

sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_xy = np.sum(X * Y)
sum_x2 = np.sum(X ** 2)

# Slope (b)
b = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)

# Intercept (a)
a = (sum_y - b * sum_x) / n

print("Intercept (a):", a)
print("Slope (b):", b)
```

```python
print("Regression Equation: Y =", a, "+", b, "* X")
```

```python
Y_pred = a + b * X
print("Predicted Values:", Y_pred)
```

```python
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Correlation and Regression")
plt.show()
```
