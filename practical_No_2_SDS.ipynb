1.What is Data Ingestion and why is it important? Which library is used?

Data ingestion is the process of collecting and importing data from different sources such as CSV files, databases, or APIs into a system for analysis. It is important in data science because without properly loading data, no analysis or machine learning can be performed. It ensures that data is available, organized, and ready for further processing. The most commonly used Python library for data ingestion and cleaning is Pandas, as it provides easy functions to load, manipulate, and clean datasets efficiently.

2.Why is data cleaning necessary before performing analysis or machine learning?

Data cleaning is necessary because raw data often contains errors such as missing values, duplicate records, incorrect formats, or irrelevant information. If this unclean data is used directly, it can lead to inaccurate analysis and poor performance of machine learning models. Cleaning the data improves its quality, ensures consistency, and helps in obtaining reliable and meaningful results.


3.How do you load a CSV file into a Pandas DataFrame?

A CSV file can be loaded into a Pandas DataFrame using the read_csv() function. First, the Pandas library is imported, and then the file is read by specifying its name or path. For example, import pandas as pd and then df = pd.read_csv('file_name.csv'). This command loads the data into a DataFrame called df, which can then be used for further analysis.


4.What is the difference between head() and tail() functions in Pandas?

The head() function is used to display the first few rows of a dataset, usually the first 5 rows by default, while the tail() function displays the last few rows of the dataset. These functions are useful for quickly understanding the structure and contents of the data without printing the entire dataset.

```python
import pandas as pd

# Reload the dataset to ensure it's not empty
df = pd.read_csv('cloud_outages_dataset.csv')

# Re-check for duplicates after reloading
initial_duplicate_count = df.duplicated().sum()
print(f"Initial number of duplicate rows: {initial_duplicate_count}")

df.drop_duplicates(inplace=True)

# Check again after dropping
duplicate_count_after_drop = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicate_count_after_drop}")
print(f"DataFrame shape after duplicate removal: {df.shape}")
```

```python
# Drop columns with all missing values
df.drop(columns=['post_mortem_link', 'previous_incident_id'], inplace=True)

# Fill missing values in 'external_factors' with 'Unknown'
df['external_factors'].fillna('Unknown', inplace=True)

print("Missing values after handling:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

print("Data types after conversion:")
display(df[['start_time', 'end_time']].info())
```

```python
# Example: Filter for incidents with 'High' severity
high_severity_incidents = df[df['severity'] == 'High']
print("Number of high severity incidents:", len(high_severity_incidents))

# Display the first 5 rows of the filtered data
display(high_severity_incidents.head())
```

```python
output_file_path = 'cloud_outages_cleaned.csv'
df.to_csv(output_file_path, index=False)
print(f"Cleaned data saved to {output_file_path}")
```
