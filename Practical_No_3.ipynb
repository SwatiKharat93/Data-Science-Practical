1.What is data wrangling? Explain the difference between data wrangling and data preprocessing.

Data wrangling means cleaning and converting raw data into a useful format so that we can analyze it easily. It includes tasks like removing errors, fixing missing values, changing data types, and organizing data.

Difference:

Data wrangling focuses on cleaning, structuring, and organizing messy raw data.
Data preprocessing is a broader step used in machine learning. It includes data wrangling plus additional steps like normalization, encoding categorical data, and feature scaling.


2.What is the importance of exploratory data analysis (EDA) in Pandas?

Exploratory Data Analysis (EDA) is an important step where we understand the dataset before applying any model. In Pandas, EDA helps us to check data structure, find missing values, detect patterns, and identify errors or outliers. It also helps us understand relationships between variables. By using functions like head(), describe(), and info(), we can quickly explore the dataset. EDA helps in making better decisions and improves the accuracy of analysis.


3.What are the main differences between Series, DataFrame, and Panel?

Series: It is a one-dimensional array (like a single column). It contains data and labels (index).
DataFrame: It is a two-dimensional table (rows and columns), like an Excel sheet.
Panel: It is a three-dimensional data structure (like multiple DataFrames). It is rarely used now and mostly removed from modern Pandas.


4.Explain the use of: a. isnull() b. notnull() c. dropna() in paragraph and easy d. fillna()

a. isnull()
This function is used to check missing values in the dataset. It returns True where data is missing and False where data is present.

b. notnull()
This is opposite of isnull(). It returns True for available values and False for missing values.

c. dropna()
This function is used to remove missing values from the dataset. It can delete rows or columns that contain null values. It helps in cleaning the data but may reduce dataset size.

d. fillna()
This function is used to replace missing values with a specific value like 0, mean, median, or any other value. It helps in keeping the dataset complete without deleting data.

5.What is column transformation and why is it important in data analysis?

Column transformation means changing the values or format of a column to make it more useful for analysis. For example, converting text data into numbers, scaling values, or creating new columns from existing ones.

It is important because raw data is often not suitable for analysis or machine learning. Transformation improves data quality, makes patterns easier to understand, and helps models perform better.

```python
import pandas as pd
```

```python
df = pd.read_csv("cloud_dataset.csv")
print(df.head())
```

```python
# Select single column
print(df['cloud_provider'])

# Select multiple columns
print(df[['cloud_provider', 'region', 'severity']])

# Select rows
print(df.loc[0:5])

# Conditional selection
print(df[df['severity'] == 'High'])
```

```python
# Create new column (example: impact score total)
df['total_impact'] = df['region_impact_score'] + df['service_impact_score']
print(df[['region_impact_score', 'service_impact_score', 'total_impact']].head())
```

```python
# Sort by outage duration
df_sorted = df.sort_values(by='duration_minutes', ascending=False)
print(df_sorted.head())
```

```python
df.rename(columns={'duration_minutes': 'outage_duration'}, inplace=True)
print(df.head())
```

```python
print(df.describe())   # numerical summary
print(df.info())       # dataset structure
```

```python
# Filter outages with high severity and SLA violation
filtered_data = df[(df['severity'] == 'High') & (df['sla_violation'] == True)]
print(filtered_data.head())
```

```python
# Average outage duration per cloud provider
grouped_data = df.groupby('cloud_provider')['outage_duration'].mean()
print(grouped_data)
```

```python
# Create another sample dataset for merging
extra_data = df[['incident_id', 'cloud_provider']].copy()

# Merge on incident_id
merged_data = pd.merge(df, extra_data, on='incident_id')
print(merged_data.head())
```

```python
# Drop a column
df.drop('notes', axis=1, inplace=True)

# Drop first row
df.drop(0, axis=0, inplace=True)

print(df.head())
```

```python
# Custom function to classify outage
def classify_outage(duration):
    if duration > 120:
        return "Critical"
    else:
        return "Normal"

df['outage_type'] = df['outage_duration'].apply(classify_outage)
print(df[['outage_duration', 'outage_type']].head())
```
