1.What is the difference between Binomial, Poisson, and Normal distributions?

Binomial Distribution
This one is obsessed with counting success vs failure. Fixed number of trials, each trial has only two outcomes.
Example: Tossing a coin 10 times and counting how many heads you get.
Key idea: fixed trials + same probability each time
Poisson Distribution
This one deals with counting events over time or space, but without a fixed number of trials.
Example: Number of customers arriving at a shop in 1 hour.
Key idea: events happening randomly at a constant rate
Normal Distribution
The famous bell curve everyone worships. Continuous data, symmetric shape, most values cluster around the mean.
Example: Heights of students in a class.

5.What is skewness in distributions and which distributions exhibit it?

Skewness means the asymmetry of a distribution.

If data is not symmetric, it is skewed.

Types of skewness:

1.Positive Skew (Right skew)

Tail is longer on the right side
Mean > Median
Example: Income distribution

2.Negative Skew (Left skew)

Tail is longer on the left side
Mean < Median
Example: Easy exam scores
Binomial Distribution
Can be skewed depending on probability (p)
Symmetric only when p = 0.5
Poisson Distribution
Positively skewed, especially when λ is small
Becomes more symmetric as λ increases
Normal Distribution
Not skewed at all
Perfectly symmetric

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm
```

```python
n = 10
p = 0.5
```

```python
x_binom = np.arange(0, n+1)
y_binom = binom.pmf(x_binom, n, p)
```

```python
plt.figure(figsize=(6,4))
plt.bar(x_binom, y_binom)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
lam = 3
```

```python
x_poisson = np.arange(0, 15)
y_poisson = poisson.pmf(x_poisson, lam)
```

```python
plt.figure(figsize=(6,4))
plt.bar(x_poisson, y_poisson, color='orange')
plt.title("Poisson Distribution (λ=3)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
mu = 0
sigma = 1
```

```python
x_norm = np.linspace(-5, 5, 100)
y_norm = norm.pdf(x_norm, mu, sigma)
```

```python
plt.figure(figsize=(6,4))
plt.plot(x_norm, y_norm, color='green')
plt.title("Normal Distribution (μ=0, σ=1)")
plt.xlabel("Values")
plt.ylabel("Density")
plt.grid()
plt.show()
```
