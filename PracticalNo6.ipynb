1.What is web scraping? Explain the role of the Requests library in web scraping.
Web scraping is the process of extracting data from websites automatically using programs or scripts. It helps to collect large amounts of data quickly without manual effort.
For example, extracting product prices, news articles, or weather information from websites.
Role of Requests Library:
The Requests library in Python is used to send HTTP requests to a website and receive the response.
It helps to access the webpage content.
It retrieves HTML data of the webpage.
It acts as a bridge between the user and the website.

2.What is BeautifulSoup? How does it help in parsing HTML?
BeautifulSoup is a Python library used for parsing HTML and XML documents.
It helps to:
Convert raw HTML into a structured format.
Easily navigate and search HTML elements.
Extract specific data like text, links, headings, etc.
How it helps in parsing:
It creates a parse tree of the HTML document.
Makes it easy to find elements using tags and attributes.
Simplifies data extraction from complex webpages.

3.Difference between Web Scraping and Web Crawling.
Web Scraping:
Web scraping is the process of extracting specific data from websites. It focuses on collecting useful information such as text, images, prices, or any required data from a webpage.
It is used for data extraction.
It targets specific information from selected web pages.
It requires tools like BeautifulSoup, Requests, or Scrapy.
Example: Extracting product prices from an e-commerce website.
Web Crawling:
Web crawling is the process of automatically browsing and navigating through web pages by following links. It is mainly used by search engines to discover and index web pages.
It is used for searching and indexing websites.
It explores multiple web pages automatically.
It is used by search engines like Google.
Example: A search engine bot visiting different pages to index content.

4.What are the commonly used methods in BeautifulSoup?
Some important methods include:
find() → Returns the first matching element
find_all() → Returns all matching elements
get_text() → Extracts text from HTML tags
select() → Uses CSS selectors to find elements
get() → Retrieves attribute values (like links)
These methods help in locating and extracting required data from web pages efficiently.

5.What are the ethical and legal considerations in web scraping?
While web scraping is useful, it must be done responsibly.
Important considerations:
Follow website rules: Always check the website’s terms and conditions and robots.txt file.
Avoid overloading servers: Sending too many requests can harm the website.
Respect privacy: Do not extract sensitive or personal data.
Use data properly: Give credit if required and avoid misuse.
Legal compliance: Some websites strictly prohibit scraping.

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Send Request to Website
url = "https://news.ycombinator.com"

response = requests.get(url)

print(response.status_code)
```

```python
# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

print(type(soup))
```

```python
# Inspect Data
print(soup.prettify()[:1000])
```

```python
# Find All Titles
titles = soup.find_all("span", class_="titleline")

print(len(titles))
```

```python
for t in titles:   # Extract and Print Data
    print(t.text)
```

```python
for t in titles: # Extract links
    link = t.find("a")["href"]
    print("Title:", t.text)
    print("Link:", link)
    print("-" * 50)
```
