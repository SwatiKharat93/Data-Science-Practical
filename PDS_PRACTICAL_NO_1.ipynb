2.What is the difference between a list and a tuple in Python?
Why would you prefer a tuple over a list in real-world applications?

Answer :A list and a tuple in Python are both sequence data types used to store multiple elements in an ordered manner, but they differ mainly in mutability. A list is mutable, which means its elements can be changed, added, or removed after creation, whereas a tuple is immutable, meaning its elements cannot be modified once defined. Lists use square brackets [], while tuples use parentheses ( ). Lists are slightly slower and consume more memory compared to tuples, whereas tuples are faster and more memory-efficient.

In real-world applications, tuples are preferred when the data should remain constant, such as storing coordinates, dates, or fixed records. They also provide better performance and can be used as keys in dictionaries due to their immutability. Therefore, tuples are chosen over lists when data safety, speed, and fixed structure are important.


3.How are Python arrays different from lists?
When should you use the array module instead of a list?

Answer: Python arrays and lists are both used to store multiple elements, but they have some important differences. A list can store elements of different data types (like integers, strings, floats), whereas an array (from the array module) can store only elements of the same data type. Lists are more flexible and easy to use, while arrays are more memory-efficient and faster when working with large amounts of numerical data. Arrays require you to specify the data type in advance, but lists do not have this restriction.
In real-world applications, you should use the array module instead of a list when you are working with a large amount of numerical data and need better performance and lower memory usage. Arrays are especially useful in mathematical operations, scientific calculations, and situations where all elements are of the same type. However, for general-purpose programming where flexibility is needed, lists are usually preferred.

4.Explain how dictionaries work internally in Python.
Why are dictionary lookups faster compared to lists or tuples?

Answer:Python dictionaries store data in key–value pairs and work internally using a hash table. When you add a key, Python applies a special function called a hash function to convert the key into a unique number (called a hash value). This hash value decides where the data will be stored in memory. So instead of searching one by one, Python directly jumps to the correct location using the key.

Dictionary lookups are faster because they use this direct access method (hashing). It usually takes constant time O(1) to find a value. In contrast, lists and tuples use sequential search, where Python may need to check each element one by one, taking more time (O(n)). That’s why dictionaries are much faster for searching data.


5.What is the difference between sets and lists in Python?
How do sets handle duplicate values, and where are sets useful in practical scenarios?

Answer : Lists and sets in Python are both used to store multiple elements, but they are quite different in behavior. A list is an ordered and mutable collection that allows duplicate values, meaning the same element can appear multiple times. A set, on the other hand, is an unordered collection that is also mutable but does not allow duplicate values. Lists use square brackets [ ], while sets use curly braces { }.

Sets automatically handle duplicate values by removing them. If you try to add the same element multiple times, it will be stored only once. This happens because sets internally use a hashing mechanism, which ensures that each element is unique.

In practical scenarios, sets are very useful when you need to work with unique data. For example, removing duplicates from a list, checking membership quickly (like whether an item exists or not), and performing mathematical operations such as union, intersection, and difference. Sets are commonly used in data analysis, filtering duplicate records, and solving problems where uniqueness is important.

```python
import array

temps = array.array('i', [])

n = int(input("Enter number of days: "))

for i in range(n):
    t = int(input("Enter temperature: "))
    temps.append(t)

avg = sum(temps) / len(temps)

print("Temperatures:", temps)
print("Average Temperature:", avg)
print("Highest:", max(temps))
print("Lowest:", min(temps))
```

```python
import array

# Student marks using array

arr = array.array('i', [])

n = int(input("Enter number of students: "))

for i in range(n):
    m = int(input("Enter marks: "))
    arr.append(m)

total = sum(arr)
avg = total / len(arr)

print("\nMarks:", arr)
print("Total:", total)
print("Average:", avg)
print("Highest:", max(arr))
print("Lowest:", min(arr))
```
