1.What is Data Science pipeline ? Expalin its key stages with example .
A Data Science pipeline is a step-by-step process used to convert raw data into meaningful insights or predictions. It ensures that data flows through different stages in a structured way. The key stages include data collection (e.g., collecting student marks data), data preprocessing (cleaning missing or incorrect values), exploratory data analysis (EDA) (understanding patterns using graphs), feature selection/engineering (choosing or creating useful variables), model building (applying algorithms like regression), evaluation (checking accuracy using metrics), and deployment (using the model in real applications). For example, predicting student marks based on study hours follows all these steps from collecting data to generating predictions.

2.What is Exploratory Data Analysis (EDA)? How do univariate and bivariate analyses help in understanding the dataset?

Exploratory Data Analysis (EDA) is the process of analyzing and visualizing data to understand its structure, patterns, and relationships before applying any model. It helps in identifying trends, outliers, and errors. Univariate analysis focuses on a single variable, such as analyzing the distribution of marks using histograms or mean values. It helps understand data spread and central tendency. Bivariate analysis involves two variables, such as study hours and marks, and helps identify relationships between them using scatter plots or correlation. Together, these analyses provide a deeper understanding of the dataset and guide further modeling.

3.Differentiate between Label Encoding and One-Hot Encoding. When should each be used?

Label Encoding converts categorical data into numerical form by assigning a unique number to each category (e.g., Red = 0, Blue = 1, Green = 2). It is simple and memory-efficient but may introduce an unintended order among categories. One-Hot Encoding, on the other hand, creates separate binary columns for each category (e.g., Red → [1,0,0], Blue → [0,1,0]). It avoids any ordinal relationship and is more suitable for nominal data. Label Encoding is best used when categories have a natural order (like low, medium, high), while One-Hot Encoding is preferred when there is no order (like colors or city names).

4.Why is feature scaling required? Compare normalization and standardization.

Feature scaling is required to ensure that all input variables are on a similar scale, especially when features have different ranges. Without scaling, variables with larger values may dominate the model and reduce performance. Normalization scales data between 0 and 1 using minimum and maximum values, making it useful when data does not follow a normal distribution. Standardization transforms data to have a mean of 0 and standard deviation of 1, making it suitable for normally distributed data. In summary, normalization is used for bounded data ranges, while standardization is preferred for statistical modeling and algorithms like regression.

5.What is feature engineering? Explain how creating new features can improve model performance.

Feature engineering is the process of creating new features or modifying existing ones to improve the performance of a machine learning model. It involves combining, transforming, or extracting useful information from raw data. For example, instead of using only study hours, we can create a new feature like “study efficiency” or categorize hours into groups. These new features can capture hidden patterns and relationships that the original data may not show clearly. By providing more meaningful inputs, feature engineering helps the model learn better, increases accuracy, and improves prediction performance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Sample dataset
data = {
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [30, 35, 50, 55, 65, 70, 75, 85]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Check missing values  # Data Preprocessing
print(df.isnull().sum())

# Define features and target
X = df[["Study_Hours"]]
Y = df["Marks"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training Data:", len(X_train))
print("Testing Data:", len(X_test))    # Train test split
```

```python
model = LinearRegression()        # Build Model
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)

print("Predicted:", Y_pred)
print("Actual:", Y_test.values)
```

```python
mse = mean_squared_error(Y_test, Y_pred) # Evaulation
r2 = r2_score(Y_test, Y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

```python
plt.scatter(X, Y)
plt.plot(X, model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Data Science Pipeline - Regression")
plt.show()
```
