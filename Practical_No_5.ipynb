2.What is the difference between Matplotlib and other visualization libraries?
Matplotlib is a fundamental data visualization library in Python that provides a wide range of plotting functions. It is considered a low-level library because it gives detailed control over every element of a graph, such as axes, labels, colors, and styles. However, this flexibility requires more code and effort to create complex visualizations.

On the other hand, other visualization libraries such as Seaborn, Plotly, and Pandas plotting are generally high-level libraries built on top of Matplotlib. These libraries are designed to simplify the process of data visualization by providing predefined styles, themes, and functions. As a result, they require less code and are easier to use, especially for beginners.

In terms of appearance, Matplotlib produces basic and simple graphs by default, whereas other libraries offer more visually appealing and modern designs. Additionally, some libraries like Plotly support interactive visualizations, allowing users to zoom, hover, and explore data dynamically, which is not a primary feature of Matplotlib.

3.What is the purpose of the pyplot module in Matplotlib?
The pyplot module in Matplotlib is used to provide a simple and convenient interface for creating different types of plots and graphs in Python. It acts as a collection of functions that allow users to generate visualizations in a step-by-step manner, similar to plotting in MATLAB.

The main purpose of pyplot is to make plotting easier by handling tasks such as creating figures, plotting data, adding titles, labeling axes, and displaying the final output.Instead of writing complex code to manage graphical elements, users can use pyplot functions to quickly create visual representations of data.

It also helps in controlling various parts of a graph, such as line styles, colors, markers, grid, and legends, making it flexible for customization. Overall, pyplot simplifies the process of data visualization and is widely used for creating basic to moderately complex plots in data analysis.


4.Explain the process of creating a basic line plot.
Creating a basic line plot in Matplotlib involves a series of simple steps that help in representing data visually.

First, the Matplotlib library is imported, specifically the pyplot module, which provides functions for plotting. After importing the library, the next step is to prepare the data. This includes defining the values for the x-axis and y-axis, usually in the form of lists or arrays.

Once the data is ready, a line plot is created using a plotting function that connects the data points with a line. After plotting the data, labels are added to the x-axis and y-axis to describe what each axis represents. A title is also given to the graph to explain the purpose of the plot.

Finally, the plot is displayed on the screen using a function that renders the graph. These steps together help in creating a simple and clear line plot that can be used to analyze trends and patterns in the data.

5.Why is customization important in data visualization? What customization options are available in Matplotlib?

Customization is important in data visualization because it improves the clarity, readability, and effectiveness of a graph. A well-customized visualization helps the audience easily understand the data and focus on important insights. Without proper customization, graphs may look confusing or fail to communicate the intended message.

Customization also helps in highlighting key information, such as trends, comparisons, or outliers, making the visualization more meaningful. It enhances the overall presentation quality and ensures that the visualization is suitable for different audiences, whether technical or non-technical.
Customization Options Available in Matplotlib

Matplotlib provides a wide range of customization options to modify the appearance of plots:

Titles and Labels: Add titles and axis labels to describe the graph clearly.
Colors: Change colors of lines, bars, and markers to make plots visually appealing.
Line Styles: Modify line types such as solid, dashed, or dotted lines.
Markers: Add symbols (like circles, squares) to highlight data points.
Legends: Include legends to explain multiple data series.
Grid: Add grid lines to improve readability.
Figure Size: Adjust the size of the plot for better visualization.
Axis Limits: Set limits for x-axis and y-axis to focus on specific data ranges.
Font Size and Style: Change font size, style, and color for better presentation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'Marks': [35, 40, 50, 55, 65, 70, 80, 90],
    'Sleep_Hours': [6, 7, 5, 6, 7, 8, 6, 7]
}

df = pd.DataFrame(data)
```

```python
plt.figure()
plt.scatter(df['Hours_Studied'], df['Marks'])
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure()
plt.plot(df['Hours_Studied'], df['Marks'])
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.title('Line Plot')
plt.show()
```

```python
plt.figure()
plt.bar(df['Hours_Studied'], df['Marks'])
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.title('Bar Plot')
plt.show()
```

```python
plt.figure()
plt.hist(df['Marks'], bins=5)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Marks'])
plt.title('Box Plot')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x = [1, 2, 3, 4]
y = [3, 4, 5, 7]

data = pd.DataFrame({
    'X': x,
    'Y': y
})

print(data)
```

```python
plt.figure()
plt.scatter(data['X'], data['Y'])
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.figure()
plt.plot(data['X'], data['Y'], marker='o')
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.figure()
plt.bar(data['X'], data['Y'])
plt.title("Bar Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
plt.figure()
plt.boxplot(data['Y'])
plt.title("Box Plot")
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```
